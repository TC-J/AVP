```python
from avp.audio.signal import ndsignal

from avp.audio.filter import iirfilter

import numpy as np

import matplotlib.pyplot as plt

from IPython.display import Audio
```

```python
# load audio
piano_c = ndsignal.load("../../assets/audios/piano_c.wav")

# create filter
filter = iirfilter(piano_c.sr, [650, 1000], btype="bandpass", plot_fftbins=piano_c.sr // 2)

plt.style.use("bmh")

plt.figure(figsize=(20, 24))

plt.subplot(6, 1, 1)

# plot signal
piano_c.plot()

_, ffts = piano_c.harmonics

plt.subplot(6, 1, 2)

# plot harmonics
piano_c.harmonic_plot()

plt.subplot(6, 1, 3)

# plot filter frequency response
filter.plot_frequency_response()

plt.subplot(6, 1, 4)

# plot filter phase response
filter.plot_phase_response()

# apply filter
piano_c_new = filter(piano_c)

plt.subplot(6, 1, 5)

# plot filtered signal
piano_c_new.plot()

plt.subplot(6, 1, 6)

# plot filtered harmonics
piano_c_new.harmonic_plot()
```
